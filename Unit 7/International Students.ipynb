### Bivariate Data Exploration

```python
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.barplot(x='Citizenship', y='Total',data=df)

plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.barplot(x='UG', y='Total',data=df)

plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.barplot(x='G', y='Total',data=df)

plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.barplot(x='ExhangeVisiting', y='Total',data=df)


plt.show()
```
